```python
#| export
from gradio_client import Client, handle_file
def predict_image(client, impath):
    """Given impath, returns a PIL annotated image, breed and confidence"""
    result = client.predict(
		img=handle_file(impath),
		api_name="/predict"
    )
    return result
```

```python
client = Client("deman539/dog_breed_id")
impath = 'https://upload.wikimedia.org/wikipedia/commons/9/9c/Yellow_Labrador_Retriever_2.jpg'
annotated_impath, breed, confidence = predict_image(client, impath)
```
